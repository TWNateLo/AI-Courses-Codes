```python
import matplotlib.pyplot as plt

listx = [0,5,7,12,18,20]
listy=[20,60,32,45,78,56]
plt.plot(listx,listy)
plt.show()
```

```python
listx = [0,5,7,12,18,20]
listy=[20,60,32,45,78,56]
plt.plot(listx,listy,color='red',lw=3,ls=':',marker='o',ms='5',label='Helllo')
plt.legend()
plt.title('Data analysis')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
listx = [0,5,7,12,18,20]
listy=[20,60,32,45,78,56]
plt.plot(listx,listy,color='green',lw=3,ls=':',marker='o',ms='5',label='Helllo')
plt.legend()
plt.title('Data analysis')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.ylim(0,200)
plt.xlim(0,100)
plt.grid(color='black',linestyle=":", linewidth=1,alpha=0.2)
plt.show()
```

```python
listx1 = [1,5,7,9,13,16]
listy1 = [15,50,80,40,70,50]
plt.plot(listx1 ,listy1 ,color="red",ls="-.",marker="o", label="Jan.")
listx2 = [15,20,25,30,35,40]
listy2 = [1,5,7,9,13,16]
plt.plot(listx2 ,listy2 ,color="blue",ls="-.",marker="o" ,label="Jan.")
plt.legend()
plt.show()
```

```python
listx =['c','c++','c#','java','python']
listy= [45,28,38,32,32]
plt.bar(listx,listy,width=0.5,color=['red','green','blue'])
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.rcParams['font.sans-serif']  = 'DFKai-SB'
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```

```python
listx =['c','c++','c#','java','python']
listy= [45,28,38,32,32]
listy1= [25,20,20,16,28]
listy2 = [20,8,18,16,22]

plt.bar(listx,listy1,width=0.5,label="")
# bottom -> rebase
plt.bar(listx,listy2,width=0.5,bottom=listy1,label="")
# legend -> to show a desciprtion small box at the top left corner
plt.legend()
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.rcParams['font.sans-serif']  = 'DFKai-SB'
plt.rcParams['axes.unicode_minus'] = False

plt.show()
```
